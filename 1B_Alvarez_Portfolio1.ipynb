### MY BUDGET TRACKER ###

DESCRIPTION:
This program calculates a student's total expenses on one day and remaining weekly balance.

```python
# the display title #
print( )
print( )
print(" ===== MY BUDGET TRACKER ===== ")
print(" Note: only input the expenses you used TODAY. ")
print( )

# user information #
student_name = input("Enter your name: ")
allowance = float(input("Enter your weekly allowance: ₱"))
food = float(input("Money spent on food: ₱"))
school = float(input("Money spent on school supplies: ₱"))
transport = float(input("Money spent on transportation: ₱"))
entertainment = float(input("Money spent on entertainment: ₱"))

# computation of total expenses #
total = food + school + transport + entertainment

# computation of remaining money #
remaining_balance = allowance - total

# the output section #
print( )
print( )
print(" --- DAILY REPORT --- ")
print( )
print("Student:",student_name)
print("Allowance: ₱",allowance)
print("Total Expenses: ₱",total)
print("Remaining Balance: ₱",remaining_balance)

# final message #
print ()
print(f'student_name, ", you have a remaining budget of ₱",remaining_balance, "this week.")
print("GREAT JOB ON MANAGING YOUR BUDGET! I'M SO PROUD OF YOU!")
```
