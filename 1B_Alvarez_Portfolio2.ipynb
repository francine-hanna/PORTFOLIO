DESCRIPTION: This program evaluates the risk level during laboratory session of student. The system evaluates the student’s laboratory condition by analyzing the number of chemicals handled, laboratory working hours, PPE compliance score, and possible laboratory equipment accidents.

```python
# the display title #
print( )
print(" ======= 🧪 LABORATORY SAFETY RISK DETECTOR 🧪 ======= ")
print( )


# user input section #
student_name = input("Enter your name: ")
chemical_count = int(input("How many chemicals did you handle today? "))
lab_hours = float(input("How many hours did you stay in the laboratory? "))
ppe_score = float(input("Enter your PPE compliance score (0-100): "))
equipment_damage = input("Did you break or crack any laboratory equipment? (yes/no): ")


# processing section (calculations) #

# chemical exposure score
chemical_score = chemical_count * 10
# laboratory fatigue score 
fatigue_score = lab_hours * 8
# initial total risk score 
risk_score = chemical_score + fatigue_score - ppe_score


# conditional excecution section #

# add additional risk points if equipment was damaged
if equipment_damage == "yes":
    risk_score = risk_score + 50
    damage_message = "Laboratory equipment incident recorded."
else:
    damage_message = "No laboratory equipment damage recorded."

# determine laboratory risk level
if risk_score >= 120:
    risk_level = " 🚨 CRITICAL LABORATORY RISK 🚨 "
    safety_message = "Immediate supervision is recommended."
elif risk_score >= 80:
    risk_level = "❗ HIGH LABORATORY RISK ❗"
    safety_message = "Additional caution is necessary."
elif risk_score >= 40:
    risk_level = "⚠️ MODERATE LABORATORY RISK ⚠️"
    safety_message = "Maintain proper laboratory discipline."
else:
    risk_level = "✅ LOW LABORATORY RISK ✅"
    safety_message = "Laboratory condition is stable."


# output section #
print( )
print( )
print(" ------ 📋 LABORATORY REPORT 📋 ------ ")
print( )
print("Student Name:", student_name)
print( )
print("Chemicals Handled:", chemical_count)
print("Laboratory Hours:", lab_hours)
print("PPE Compliance Score:", ppe_score)
print( )
print("Equipment Damage Status:", equipment_damage)
print("Incident Report:", damage_message)
print( )
print("Computed Risk Score:", risk_score)
print( )
print("Risk Level:", risk_level)
print("Safety Evaluation:", safety_message)


# final message #
print( )
print(" ----- 💥 MISSION COMPLETE! 💥 ------")
print("You survived another laboratory session,", student_name + "!")
print("Keep your goggles on and your beakers unbroken.")
```
