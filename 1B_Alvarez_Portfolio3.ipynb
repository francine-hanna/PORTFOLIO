DESCRIPTION:

    "PagePulse: Smart Reading Companion" is a personalized Python program designed for readers who enjoy tracking their reading progress, discovering book genres based on mood, and staying motivated while reading. The system uses functions to organize different tasks into reusable sections while creating an interactive and enjoyable experience for users.

```python
# FUNCTION 1:
# display program header #

def display_header():

    print( )
    print( )
    print("======== 📚 PagePulse 📚 ========")
    print( )
```

```python
# FUNCTION 2:
# calculate the reading percentage #

def calculate_reading_percentage(read, total):
    return (read / total) * 100
```

```python
# FUNCTION 3:
# this function determines the reader's reading level based on progress #

def determine_reading_status(reading_percentage):
      
    if reading_percentage >= 100:
        reading_status = "🏆 Book Master"
    elif reading_percentage >= 75:
         reading_status = "📖 Dedicated Reader"
    elif reading_percentage >= 50:
         reading_status = "✨ Consistent Reader"
    elif reading_percentage >= 25:
         reading_status = "🥀 Growing Book Explorer"
    else:
         reading_status = "🌱 Beginner Reader"
 
    return  reading_status
```

```python
# FUNCTION 4:
# this function recommends a book genre based on the user's mood #

def recommend_book_genre(user_mood):
    
    if user_mood == "happy":
        recommended_genre = "🧙 Fantasy"
    elif user_mood == "sad":
        recommended_genre = "😂 Comedy"
    elif user_mood == "stressed":
        recommended_genre = "🌸 Slice of Life"
    elif user_mood == "curious":
        recommended_genre = "🕵 Mystery"
    else:
        recommended_genre = "🚀 Adventure"

    return recommended_genre
```

```python
# FUNCTION 5:
# this function generates a motivational reading message based on reading status #

def generate_motivation_message(reading_status):

    if reading_status == "🏆 Legendary Book Finisher":
        motivation_message = ("🌟 Your dedication to reading is inspiring! 🌟")
    elif reading_status == "📖 Elite Page Turner":
        motivation_message = ("🔥 You are getting closer to finishing the book! 🔥")
    elif reading_status == "✨ Focused Reader":
        motivation_message = ("📚 Your reading progress is improving steadily. 📚 ")
    elif reading_status == "🌱 Growing Book Explorer":
        motivation_message = ("💡 Every page you read builds knowledge. 💡 ")
    else:
        motivation_message = ("✨ Every great reader starts with one page. ✨")

    return motivation_message
```

```python
# FUNCTION 6:
# this function displays the final personalized reading report #

def display_final_report(
    reader_name,
    book_title,
    reading_percentage,
    reading_status,
    recommended_genre,
    motivation_message):

    print( )
    print(" -------- 📖 FINAL READING REPORT 📖 --------")
    print( )

# display reader information 
    print("Reader Name:", reader_name)
    print("Current Book:", book_title)

# display progress information #
    print( )
    print("Reading Percentage:", reading_percentage)
    print("Reading Status:", reading_status)

# display recommendation #
    print( )
    print("Recommended Genre:", recommended_genre)

# display motivation #
    print( )
    print("Motivational Message:")
    print(motivation_message)

# personalized closing message #
    print( )
    print("Keep enjoying your reading journey,", reader_name + "!")
```

```python
# MAIN PROGRAM SECTION
# this section controls the flow of the entire program #

# call the header function #
display_header()


# USER INPUT SECTION
reader_name = input("Enter your name: ")
book_title = input("Enter the title of your book: ")
total_pages = int(input("Enter the total number of pages: "))
pages_read = int(input("Enter the number of pages read: "))
user_mood = input("Enter your mood (happy/sad/stressed/curious): ")


# PROCESSING SECTION
# calculate reading percentage #
final_reading_percentage = (
    calculate_reading_percentage(
        pages_read,
        total_pages))

# determine reading status #
final_reading_status = (
    determine_reading_status(
        final_reading_percentage))

# recommend a book genre #
final_recommended_genre = (
    recommend_book_genre(
        user_mood))

# generate motivation message
final_motivation_message = (
    generate_motivation_message(
        final_reading_status))

# OUTPUT SECTION
# call the report function #
display_final_report(
    reader_name,
    book_title,
    final_reading_percentage,
    final_reading_status,
    final_recommended_genre,
    final_motivation_message)


# FINAL MESSAGE SECTION
print( )
print("✅ PROGRAM COMPLETED SUCCESSFULLY ✅")
print("🌟 Thank you for using PagePulse! 🌟")
```
